```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Superstore_Data_Cleaned.csv')
data.head()
```

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'])
data.info()
```

```python
data['Order_Year'] = pd.DatetimeIndex(data['Order_Date']).year
data['Order_Month'] = pd.DatetimeIndex(data['Order_Date']).month
data.head()
```

```python
data_summary = data.groupby(by=['Order_Year','Order_Month','Region','Category','Sub-Category'], dropna=False, as_index=False)['Sales','Quantity','Profit','Discount'].sum()
data_summary.head()
```

```python
data_summary.sort_values(by=['Order_Year','Order_Month'], ascending = False)
```

```python
data_month_aging = data_summary[['Order_Year','Order_Month']].drop_duplicates(subset=['Order_Year','Order_Month']).sort_values(by=['Order_Year','Order_Month'], ascending = False).reset_index(drop = True)
```

```python
data_month_aging['Month_Aging'] = data_month_aging.index

data_month_aging
```

```python
data_summary_aging = data_summary.merge(data_month_aging, on=['Order_Year','Order_Month'])
data_summary_aging.head()
```

```python
data_summary_aging1 = data_summary_aging[data_summary_aging['Month_Aging'].isin([0,1,12])]
```

```python
data_summary_aging1
```

```python
data_summary_aging1.groupby(by=['Month_Aging','Region'], dropna=False, as_index=False).sum()
```

```python
summary_pivot = pd.pivot_table(data_summary_aging1, values =['Sales','Quantity','Profit','Discount'], index =['Region','Month_Aging'], aggfunc = np.sum).transpose()
```

```python
summary_pivot.to_csv('table_pivot.csv')
```
